```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

x1 = np.random.normal(0,1,100)
y1 = np.random.normal(0,1,100)
x2 = np.random.normal(4,1.5,50)
y2 = np.random.normal(4,1.5,50)

plt.scatter(x1, y1, color = "r",label = "class1")
plt.scatter(x2, y2, color = "b",label = "class2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.title("scatter plot")
plt.show()

```
